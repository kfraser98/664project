```python
results =[]
for year in range(2015, 2026):
    yearly_data = box_office.get_yearly(year)
    top_10 = yearly_data[:10]

    for movie in top_10:
       results.append({
           'year': year,
           'rank': movie.get('Rank'),
           'title': movie.get('Release'),
           'total_gross': movie.get('Total Gross')
       })

df = pd.DataFrame(results)
print(df)
```

```python
df.to_csv('boxoffice.csv')
```
